 ### Pergunta: Será que os níveis de corrupção, assim como o índice de Mortalidade Infantil de um país, reflete seu índice de IDH?
 
 
 A corrupção em um país normalmente é responsável pelo desvio de grandes quantidades de dinheiro, pelo gasto do dinheiro público com serviços de baixa prioridade(como por exemplo na questão da Saúde Pública) e principalmente por alterar os principais focos de investimento do país. Isso acaba por alterar de alguma maneira o IDH de um país. O intuito deste trabalho é verificar se os níveis de corrupção de um país, assim como o Índice de Mortalidade Infantil deste país, refletem seus índices de IDH(ou seja, verificar se países com altos índices de corrupção e de mortalidade infantil tem índices de IDH menores, e vice-versa).
 
 As três variáveis escolhidas foram:
 
  1)IDH: É calculado a partir da Expectativa de Vida, anos médios de estudo e anos esperados de escolaridade, Renda (GDP ou PIB) per capita de um país. Escolhida por representar um país nas questões básicas da sociedade, como Sáude e Educação.
  
  2)Índice de percepção de corrupção (CPI em inglês): Índica o nível de corrupção em um país. Valores altos representam menos índice de corrupção. Foi escolhido pois a corrupção de um país, como explicado na introdução, acaba normalmente desviando o dinheiro e o foco público de questões de alta prioridade.
  
  3)Índice de Mortalidade Infantil(por 1000 crianças): Indica quantas crianças entre zero e cinco anos(0-5) morrem dentre mil  dessas crianças nascidas. Este índice foi escolhido por representar uma falta (ou não) de investimento e estrutura na Saúde Pública. O índice teve como foco crianças, que são inicialmente mais propensas a representarem problemas de saúde por terem um sistema imunológico mais fraco(devido a sua idade), necessitando assim de maiores investimentos em educação para combater tal mortalidade infantil. Deste modo, quanto maior este índice, maior a falta de investimento e estrutura na Saúde Pública.


Começaremos iniciando os módulos gráficos e bibliotecas.
 

  
  
  

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import seaborn as sns
```

Antes de iniciar a análise do conteúdo, temos de gerar os dataframes que usaremos. Iniciaremos gerando o dataframe de IDH dos
países para o ano base escolhido, que é 2015 no caso deste projeto.Para que possamos obter um diferença razoável no IDH dos diferentes países, possibilitando nossa análise, pegaremos 30 países para a análise do projeto, 15 deles com valores mais altos de IDH, e 15 com valores baixos de IDH

OBS: O IDH varia de 0 até 1

```python
#chamando o arquivo xlsx-excel

IDH= pd.read_excel("hdi_human_development_index.xlsx")

IDH.loc[:,['geo',2015]]



#renomeando colunas e linhas
IDH=IDH.loc[ :,["geo", 2015]]
IDH.rename(columns={2015:"IDH","geo":"País"},inplace=True)
IDH.set_index("País",inplace=True)

#gerando os países para a tabela

b10= IDH.tail(15)
c10= b10.append(IDH.head(15))

c10= c10.sort_values(by="IDH")
c10










```

Agora devemos criar o Dataframe do Índice de corrupção dos 30 países apresentados no Dataframe de IDH

```python
#chamando o arquivo xlsx-excel

IC= pd.read_excel("corruption_perception_index_cpi.xlsx")

IC.loc[:,['geo',2015]]

#renomeando colunas e linhas

IC=IC.loc[ :,["geo", 2015]]
IC.rename(columns={2015:"Índice de corrupção","geo":"País"},inplace=True)
IC.set_index("País",inplace=True)

#gerando os países para a tabela

d10= IC.tail(15)
e10= d10.append(IC.head(15))

e10= e10.sort_values(by="Índice de corrupção")
e10




```

Antes de iniciarmos as análises, devemos gerar o DataFrame do Índice de Mortalidade Infantil(a cada 1000 crianças de 0 a 5 anos) por país analisado.

```python
#chamando o arquivo xlsx-excel

MI= pd.read_excel("child_mortality_0_5_year_olds_dying_per_1000_born.xlsx")

MI.loc[:,['geo',2015]]

#renomeando colunas e linhas

MI=MI.loc[ :,["geo", 2015]]
MI.rename(columns={2015:"Mortalidade Infantil","geo":"País"},inplace=True)
MI.set_index("País",inplace=True)


#gerando os países para a tabela
f10= MI.tail(15)
g10= f10.append(MI.head(15))

g10= g10.sort_values(by="Mortalidade Infantil")
g10
```

Para iniciarmos a análise dos DataFrames criados, utilizaremos um tipo de estratégia para definir a proporcionalidade dos 3 índices escolhidos. Esta estratégia é composta por criar inicialmente gráfios de análise individual dos índices, e depois criaremos gráficos de análise de dois índices juntos. Deste modo, se por exemplo, o IDH e o CPI forem proporcionais, e o IDH e o índice de mortalidade infantil também forem proporcionais, isto significa que os três índices são proporcionais. Deste modo, poderemos avaliar a questão a ser respondida e chegar a um resultado mais preciso.


Começaremos analisando o IDH apenas, através do gráfico Box-Plot.

```python
#gerando o Box-Plot para o IDH

c10.plot.box();
```

Analisando-se o gráfico, vemos que a mediana do IDH dos países listados está em torno de 0,76. Isso significa que eles estão na categoria média de IDH, segundo o relatório de desenvolvimento humano. 

```python
#gerando o Box-Plot para o Índice de  percepção de corrupção

e10.plot.box();
```

A partir do gráfico BoxPlot do Índice de corrupção, dos países listados acima, podemos ver que a mediana dos países(nesse índice) é de aproximadamente 30. Sabendo que valores mais altos desse índice indicam menos corrupção no país(varia de 0 a 100), podemos tirar como conclusão que grande parte dos países apresentam índices maiores de corrupção. 

Faremos o mesmo procedimento para o índice de mortalidade Infantil.

```python
#gerando o Box-Plot para o Índice de mortalidade infantil.


g10.plot.box();
```

 Analisando-se esse gráfico, podemos dizer que a mediana desse índice está em torno de 17. Isso significa que muitos países estão com taxas elevadas de mortalidade infantil. Para podermos comparar, o país com o índice mais baixo é Andorra, com 2,8. Deste modo, vemos que 17 é um índice um tanto elevado para a mediana dos países listados acima.

##### Conclusão Inicial:

Juntando as informações dos gráficos, podemos inferir que a maioria dos países estão com um índice de desenvolvimento humano médio, aliado de altos índices de corrupção e índices médio-elevados de Mortalidade Infantil. A priori, a suposição de que um IDH baixo impacta tanto o índice de corrupção como o de mortalidade infantil parece ser verdadeira. 

Analisaremos através de tabelas e gráficos de dois índices juntos, para que possamos tirar conclusões certas e mais realistas.

Iniciaremos criando o Data Frame IDH x Índice de corrupção x Mortalidade Infantil:



```python
# Gerando o Data-Frame dos três índices juntos.


junto=[c10,e10,g10]
first=junto.pop(0)

df=first.join(junto,how="inner")

df
```

Seguindo o procedimento, analisaremos agora a correlação entre o IDH e o Índice de corrupção, para que possamos verificar a hipótese do projeto.

```python
# Plotando o gráfico de correlação entre o IDH e o Índice de corrupção

sns.regplot(df["IDH"],df["Índice de corrupção"], color=('green'))
plt.grid()
plt.title('IDH x Índice de corrupção')
plt.grid(True)
plt.show()


# Calculando a correlação entre o IDH e o Índice de corrupção

corr1 = df['IDH'].corr(df["Índice de corrupção"])
print('Correlação entre as varíaveis:'+str(corr1))
```

O gráfico acima nos mostra que o IDH e o Índice de percepção de corrupção são diretamente proporcionais. Isto significa que um aumento no IDH implica um Índice de corrupção mais baixo(quanto maior o Índice de corrupção, menos corrupto é o país). 

Uma hipótese que explica esta relação é que países com IDH mais elevado, tendem  a terem os serviços básicos, educação e segurança com maiores qualidades. Deste modo, tendo os serviços básicos com maior qualidade, a corrupção acaba por decair, por ser mais díficil  de existir algo que impulsione o crescimento da corrupção( como fiscalização fragilizada, instituições ineficientes, entre outros).

A correlação entre as variáveis é de 0,786. Por ser uma correlação de valor elevado, isto significa que ambos índices tem uma relação muito próxima quantitativamente, e isto reforça a proporcionalidade direta de ambos índices.


Analisaremos agora a correlação entre o IDH e o Índice de motalidade infantil

```python
# Plotando o gráfico de correlação entre o IDH e o Índice de Mortalidade Infantil.

sns.regplot(df["IDH"],df["Mortalidade Infantil"], color=('green'))
plt.grid()
plt.title('IDH x Mortalidade Infantil')
plt.grid(True)
plt.show()


# Calculando a correlação entre o IDH e o Índice de Mortalidade Infantil.

corr1 = df['IDH'].corr(df["Mortalidade Infantil"])
print('Correlação entre as varíaveis:'+str(-corr1))
```

A partir da análise do gráfico, verificamos que o aumento no IDH provoca um decaimento linear na taxa de Mortalidade Infantil. Esta afirmação pode ser verificada também pela correlação entre os índices analisados, cujo valor é de 0,885. Tal valor é elevado o suficiente para podermos dizer que ambos índices fazem parte de um sistema de causa e consequência, ou seja, são diretamente proporcionais.

Uma hipótese que explica tal análise, é de que a mortalidade infantil varia conforme os níveis de saúde pública oferecida pelo estado principalmente. Isto faz com que uma menor taxa de mortalidade infantil signifique um maior investimento estatal no que tange ao aspecto da saúde pública, e vice-versa. Deste modo, sabendo-se que um dos Índices do IDH é a esperança de vida, podemos analisar que uma maior esperança de vida indica maior investimento na questão de saúde pública, estando inserida neste contexto, a saúde pública direcionada às crianças, impactando assim diretamente o índice de mortalidade infantil.


Por último, antes de tirarmos uma conclusão sobre a questão analisa no projeto, analisaremos a relação entre o Índice de percepção de corrupção e o Índice de mortalidade infantil. Podemos previamente, a partir dos gráficos anteriores, dizer que eles são diretamente proporcionais também, porém para termos certeza da precisão, torna-se necessário esta análise individual entre ambos índices.

```python
# Plotando o gráfico de correlação entre o Mortalidade Infantil e o Índice de corrupção

sns.regplot(df["Índice de corrupção"],df["Mortalidade Infantil"], color=('green'))
plt.grid()
plt.title('Índice de corrupção x Mortalidade Infantil')
plt.grid(True)
plt.show()


# Calculando a correlação entre o Mortalidade Infantil e o Índice de corrupção

corr1 = df['Índice de corrupção'].corr(df["Mortalidade Infantil"])
print('Correlação entre as varíaveis:'+str(-corr1))
```

Analisando-se o gráfico, podemos ver que ambos índices são proporcionais entre si. Entretanto,  eles não apresentam uma correlação tão elevada quanto ás outras medidas( 0,683), apesar de ainda ser uma correlação de alto valor. Isto pode significar que tratando ambos índices como Causas, podemos ver que eles são relacionados de maneira linear proporcionalmente. Ampliando o espectro, podemos dizer que as consequências de ambos índices, se analisadas do mesmo modo como feito neste gráfico(juntas), provavelmente terão uma correlação mais elevada, tendendo a um índice muito elevado de correlação. Isto deve-se ao fato de que as consequências tendem a um mesmo "ponto", iniciando deste modo uma aproximação cada vez maior entre as consequências desses índices conforme ampliamos e espectro analisado.

## Conclusão:




Analisando-se todas informações obtidas previamente a partir dos gráficos e dos índices de correlação obtidos, podemos dizer que o IDH está diretamente relacionado, ou seja, tem um valor elevado de correlação entre o Índice de percepção de corrupção e o Índice de mortalidade infantil. Isto significa que a conclusão inicial estava correta e apontava para as informações corretas. Por fim, respondendo à pergunta proposta por este projeto: Sim, o Índice de corrupção e o de mortalidade infantil refletem os valores do IDH de um país, sendo que a melhor descrição entre a relação destes é de proporcionalidade linear e direta. 

Entretanto, deve-se entender que não basta analisar apenas um dos índices( Mortalidade Infantil e o de Corrupção) para que se possa tirar uma conclusão acerca do IDH do país analisado, deve-se analisar conjuntamente estes dois que para que assim, seus valores reflitam o valor de IDH do país de maneira correta e mais realista possível.
