 ### Pergunta: Será que os níveis de corrupção, assim como o índice de Mortalidade Infantil de um país, refletem seus índices de IDH?
 
 
 A corrupção em um país normalmente é responsável pelo desvio de grandes quantidades de dinheiro, pelo gasto do dinheiro público com serviços de baixa prioridade(como por exemplo na questão da Saúde Pública) e principalmente por alterar os principais focos de investimento do país. Isso acaba por alterar de alguma maneira o IDH de um país. O intuito deste trabalho é verificar se os níveis de corrupção de um país, assim como o Índice de Mortalidade Infantil deste país, refletem seus índices de IDH(ou seja, verificar se países com altos índices de corrupção e de mortalidade infantil tem índices de IDH menores, e vice-versa).
 
 As três variáveis escolhidas foram:
 
  1)IDH: É calculado a partir da Expectativa de Vida, anos médios de estudo e anos esperados de escolaridade, Renda (GDP ou PIB) per capita de um país. Escolhida por representar um país nas questões básicas da sociedade, como Sáude e Educação.
  
  2)Índice de percepção de corrupção (CPI em inglês): Índica o nível de corrupção em um país. Valores altos representam menos índice de corrupção. Foi escolhido pois a corrupção de um país, como explicado na introdução, acaba normalmente desviando o dinheiro e o foco público de questões de alta prioridade.
  
  3)Índice de Mortalidade Infantil(por 1000 crianças): Indica quantas crianças entre zero e cinco anos(0-5) morrem dentre mil  dessas crianças nascidas. Este índice foi escolhido por representar uma falta (ou não) de investimento e estrutura na Saúde Pública. O índice teve como foco crianças, que são inicialmente mais propensas a representarem problemas de saúde por terem um sistema imunológico mais fraco(devido a sua idade), necessitando assim de maiores investimentos em educação para combater tal mortalidade infantil. Deste modo, quanto maior este índice, maior a falta de investimento e estrutura na Saúde Pública.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

Antes de iniciar a análise do conteúdo, temos de gerar os dataframes que usaremos. Iniciaremos gerando o dataframe de IDH dos
países para o ano base escolhido, que é 2015 no caso deste projeto.Para que possamos obter um diferença razoável no IDH dos diferentes países, possibilitando nossa análise, pegaremos 30 países para a análise do projeto, 15 deles com valores mais altos de IDH, e 15 com valores baixos de IDH

OBS: O IDH varia de 0 até 1

```python
#chamando o arquivo xlsx-excel

IDH= pd.read_excel("hdi_human_development_index.xlsx")

IDH.loc[:,['geo',2015]]



#renomeando colunas e linhas
IDH=IDH.loc[ :,["geo", 2015]]
IDH.rename(columns={2015:"IDH","geo":"País"},inplace=True)
IDH.set_index("País",inplace=True)

#gerando os países para a tabela

b10= IDH.tail(15)
c10= b10.append(IDH.head(15))

c10= c10.sort_values(by="IDH")
c10










```

Agora devemos criar o Dataframe do Índice de corrupção dos 30 países apresentados no Dataframe de IDH

```python
#chamando o arquivo xlsx-excel

IC= pd.read_excel("corruption_perception_index_cpi.xlsx")

IC.loc[:,['geo',2015]]

#renomeando colunas e linhas

IC=IC.loc[ :,["geo", 2015]]
IC.rename(columns={2015:"Índice de corrupção","geo":"País"},inplace=True)
IC.set_index("País",inplace=True)

#gerando os países para a tabela

d10= IC.tail(15)
e10= d10.append(IC.head(15))

e10= e10.sort_values(by="Índice de corrupção")
e10




```

Antes de iniciarmos as análises, devemos gerar o DataFrame do Índice de Mortalidade Infantil(a cada 1000 crianças de 0 a 5 anos) por país analisado.

```python
#chamando o arquivo xlsx-excel

MI= pd.read_excel("child_mortality_0_5_year_olds_dying_per_1000_born.xlsx")

MI.loc[:,['geo',2015]]

#renomeando colunas e linhas

MI=MI.loc[ :,["geo", 2015]]
MI.rename(columns={2015:"Mortalidade Infantil","geo":"País"},inplace=True)
MI.set_index("País",inplace=True)


#gerando os países para a tabela
f10= MI.tail(15)
g10= f10.append(MI.head(15))

g10= g10.sort_values(by="Mortalidade Infantil")
g10
```
